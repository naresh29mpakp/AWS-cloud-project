# Welcome to Virtual-Cloth-Assistant

> It'll take some extra time in first execution for setting up and downloading of model weights

```python
def make_dir():
  os.system("cd /content/ && mkdir inputs && cd inputs && mkdir test && cd test && mkdir cloth cloth-mask image image-parse openpose-img openpose-json")
```

```python
from PIL import Image
import os

def run(cloth, model):
    make_dir()  # Ensure required directories exist
    cloth.save("/content/inputs/test/cloth/cloth.jpg")
    model.save("/content/inputs/test/image/model.jpg")

    # Remove old outputs and run the prediction script
    os.system("rm -rf /content/output/")
    os.system("python /content/clothes-virtual-try-on/run.py")

    # Check if any output is generated
    output_files = os.listdir("/content/output") if os.path.exists("/content/output") else []
    if not output_files:
        print("Error: No output files found! Check if run.py is working correctly.")
        return None  # Avoid crashing

    # Load and return the generated output image
    op = Image.open(f"/content/output/{output_files[0]}")
    return op
```

```python
import gradio as gr

def safe_run(cloth, model):
    result = run(cloth, model)
    if result is None:
        return None  # Instead of returning a string, return None to prevent Gradio from crashing
    return result

with gr.Blocks() as app:
    with gr.Row():
        gr.Markdown("<center><h1> Clothes Virtual Try-On </h1></center>")

    with gr.Row():
        with gr.Column():
            cloth_input = gr.Image(type="pil", label="Upload the Cloth Image")
        with gr.Column():
            model_input = gr.Image(type="pil", label="Upload the Human Image")

    with gr.Row():
        final_output = gr.Image(type="pil", label="Final Prediction")

    with gr.Row():
        submit_button = gr.Button("Submit")

    submit_button.click(fn=safe_run, inputs=[cloth_input, model_input], outputs=[final_output])

app.launch(debug=True, share=True)
```
